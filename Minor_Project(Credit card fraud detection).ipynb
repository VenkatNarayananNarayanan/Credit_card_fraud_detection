```python
from sklearn.preprocessing import StandardScaler
std_slr = StandardScaler()
df['std_amount'] = std_slr.fit_transform(df['Amount'].values.reshape(-1,1))
df['std_time'] = std_slr.fit_transform(df['Time'].values.reshape(-1,1))
```

# Evaluating the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.80)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr = LogisticRegression()
rfc = RandomForestClassifier()
svm = SVC()
knn = KNeighborsClassifier()
```

```python
lr.fit(x_train,y_train)
```

```python
rfc.fit(x_train,y_train)
```

```python
svm.fit(x_train,y_train)
```

```python
knn.fit(x_train,y_train)
```

```python
list(x_train[0])
```

```python
list(y_train[0])
```

```python
y_pred_lr = lr.predict(x_test)
y_pred_rfc= rfc.predict(x_test)
y_pred_svm = svm.predict(x_test)
y_pred_knn = knn.predict(x_test)
```

# Calulating the accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
lr_acc = round(accuracy_score(y_pred_lr,y_test)*100,2)
rfc_acc = round(accuracy_score(y_pred_rfc,y_test)*100,2)
svm_acc = round(accuracy_score(y_pred_svm,y_test)*100,2)
knn_acc = round(accuracy_score(y_pred_knn,y_test)*100,2)
```

```python
print("Accuracies: ")
print("Logistic Regression:",lr_acc)
print("Random Forest:",rfc_acc)
print("SVM:",svm_acc)
print("KNN:",knn_acc)
```
